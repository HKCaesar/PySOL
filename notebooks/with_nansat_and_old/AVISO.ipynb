```python
# OPeNDAP client - http://pydap.org
from pydap.client import open_url
import datetime
# Math package
import numpy as np
# Visualization package
import matplotlib.pyplot as plt
# maps toolkit
from mpl_toolkits.basemap import Basemap

# file  
link = "http://naiad-srv1.solab.rshu.ru:8080/opendap/production/AVISO_NRT_MADT_MERGED_H/nrt_global_merged_madt_h_20110120_20110120_20110126.nc.gz"

# read data
dataset = open_url(link) 
fill_value = dataset.Grid_0001.attributes['_FillValue']
lat = dataset.NbLatitudes[:]
lon = dataset.NbLongitudes[:]
madt = dataset.Grid_0001[:,:]
madt = np.transpose(madt)#[::-1,::]
madt = np.ma.masked_where(isnan(madt), madt) 
madt = np.ma.array(madt, mask = np.where(madt == fill_value, 1, 0))
x,y = np.meshgrid(lon, lat)

plt.close('all')
fig = plt.figure(figsize=(10,10))
#fig = plt.figure()
ax = axes([0.,0.,1.,1.])
ax.set_axis_off()
m = Basemap(ax=ax, projection='stere', width=10000000, height=10000000, lon_0=-45, lat_0=90, lat_ts=70, rsphere=(6378137.0, 6356752.3142), resolution='l')
#print m.proj4string
#print "+proj=stere +lat_0=90 +lat_ts=70 +lon_0=-45 +k=1 +x_0=0 +y_0=0 +ellps=WGS84 +datum=WGS84 +units=m +no_defs"
#m.drawcoastlines(linewidth=0.25)
#m.fillcontinents(color='#000000', lake_color='#666666')
#m.drawmapboundary(fill_color='#666666')

px, py = m(x,y)
m.pcolor(px, py, madt, vmin=-150,vmax=150)

#ax.set_axis_off()
#fig.add_axes(ax)  
#fig.savefig('aviso/aviso_test.png', transparent=True) 
    
#!mv aviso/*.png /media/SOLabNFS/http/public/ilya/aviso
```
