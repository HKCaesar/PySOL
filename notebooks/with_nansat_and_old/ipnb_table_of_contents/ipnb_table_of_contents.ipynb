function toggle() {
     $('#divToToggle').toggle();
     if ($('#divToToggle').css('display') == 'none') {
          $('#divToToggle').css('display') == '';
          $('#aTag').find("span").html('&#9658;');
     }
     else {
          $('#aTag').find("span").html('&#9650;');
     }
}

</script>

```python
Hiding menu with animation using CSS3 in a markdown cell

<div id="nav">
<b>Contents</b>
<a href="#" class="show">[show]</a>
<a href="#" class="hide">[hide]</a>
<ol class="nested" id="menu">
    <li class="nested"><a href="#Introduction" target="">Introduction</a></li>
    <li class="nested"><a href="#How_to_install">How to install</a>
        <ol class="nested">
            <li class="nested"><a href="#Installing_Nansat">Installing Nansat</a>
            <ol class="nested">
                <li class="nested"><a href="#Download_Nansat">Download Nansat</a></li>
                <li class="nested"><a href="#Requirements">Requirements</a></li>
                <li class="nested"><a href="#Installation_process">Installation process</a></li>
            </ol>
            </li>
            <li class="nested"><a href="#Installing_IPython_Notebook">Installing IPython Notebook</a></li>
        </ol>
    </li>
    <li class="nested"><a href="#Nansat_Capabilities_Examples">Nansat Capabilities Examples</a>
    <li class="nested"><a href="#Summary">Summary</a>
    <li class="nested"><a href="#References">References</a>
    <li class="nested"><a href="#Quiz_Answers">Quiz Answers</a>
    <li class="nested"><a href="#Abbreviations">Abbreviations</a>
</ol>
</div>

<style>

    * {font-size: 100%; font-family: "liberation sans", arial, sans-serif; color: black; }
    #nav { position: fixed; right: 0; top: 7%; z-index: 5; padding: 1em; border: 2px solid #aaa; background-color: #eef; width: auto;}
    #nav:hover { border: 2px solid #ddd; background-color: #ccf; }
    
    .hide, .show  { display: none;  font-size: 80%; vertical-align: middle; }
    
    @media all and (min-width:1px) { .hide, .show  { display: inline; } }

    /*
    Let's create two reasonably-named classes,
    .hide and .show, and say that since the box is hidden,
    we don't want the .hide class to be displayed: a.hide { display: none; }
    Next we need some rules that show the
    [show] link and hide the [hide] link when the [hide] link is clicked on.
    This can be done using the :focus pseudo-class
    */

    a.show:focus + .hide { display: inline; }
    a.hide { display: none; }
    a.show:focus { display: none; }

    #menu {
        -o-transition: all 1.5s;-moz-transition: all 1.5s;-webkit-transition: all 1.5s;
        height:0em; opacity: 0; margin-top: 0em; margin-left: -1ex; overflow: hidden;
    }
    
    a.show:focus ~ #menu { opacity: 1; height: 17em; }
    /* a.hide:focus ~ #menu { opacity: 0; height: 0em; } */
    
    @media all and (min-width:1px) { .hide, .show  { display: inline; } }
    
    @media print {
    .hide, .show  { display: none; } 
    #nav {position: static; border: none; width: auto;}
    #menu {width: auto;} 
    }
    
    ol.nested {
        counter-reset: item;
        list-style: none;
    }
    li.nested {
        display: block;
    }
    li.nested:before {
        counter-increment: item;
        content: counters(item, ".")" ";
    }

</style>
```
