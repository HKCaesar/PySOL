```python
import netCDF4 as nc
import pyresample as pr
import matplotlib as mp
mp.use("Agg") # to avoid some DISPLAY issues on compute nodes, we do not need display anyways
from matplotlib.pylab import *
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import os, sys
import mpl_toolkits.basemap as bm
from dateutil import parser
import glob
```

```python
#area_region_id = 'epsg_3413_crude'
area_region_id = 'pc_world'
#area_region_id = 'my_ortho_crude'

area_def = pr.utils.load_area('pyresample/areas.cfg', area_region_id)

ds_filename = '/media/SOLabNFS2/tmp/SOLAB_AMSRE_L2_NN/2011/2011.09/SOLAB_AMSRE_L2_NN_20110901_034055_20110901_043056_209_A_v1.nc'

ds = nc.Dataset(ds_filename,'r')

data_variable_name = 'sea_surface_temperature'
lons_variable_name = 'longitude'
lats_variable_name = 'latitude'
start_date = ds.getncattr('start_date')
start_time = ds.getncattr('start_time')
start_datetime = start_date+' '+start_time
dt = parser.parse(start_datetime)

data = ds.variables[data_variable_name][:]
lons = np.ma.MaskedArray(ds.variables[lons_variable_name][:])
lats = np.ma.MaskedArray(ds.variables[lats_variable_name][:])

swath_def = pr.geometry.SwathDefinition(lons, lats)

result = pr.kd_tree.resample_nearest(swath_def, data.ravel(),
  area_def, radius_of_influence=20000, fill_value=None, nprocs=4)

plt.close('all')
fig = figure(num = 1, figsize=(13.,13.), facecolor=None, edgecolor=None, frameon=False)

#fig = figure(num = 1, figsize=(10.,10.), facecolor='white')
ax = axes([0.,0.,1.,1.])
#ax = fig.add_axes([0.05,0.01,0.9,0.95],axisbg='black')
ax.set_axis_off()
fig.add_axes(ax)

m2 = pr.plot.area_def2basemap(area_def, resolution='c')
#m2 = pr.plot.area_def2basemap(area_def)

palette  = cm.jet
palette.set_bad('grey', 0)
pal_norm = colors.Normalize(vmin = 0, vmax = 30, clip = False)

m2.drawcoastlines(linewidth=0.5, color='black', antialiased=1, ax=None, zorder=None)
m2.drawcountries(linewidth=0.5, color='black', antialiased=1, ax=None, zorder=None)
m2.fillcontinents(color='0.9', lake_color='blue', ax=None, zorder=None)
m2.drawmapboundary(color='black', linewidth=1.0, fill_color='blue', zorder=None, ax=None)
m2.drawmeridians(np.arange(0,360,30))
m2.drawparallels(np.arange(-90,90,30))
ax.set_title('SOLAB_AMSRE_L2_NN SST, ${}^\circ{}$C\nDatetime: %s' % dt.strftime('%d-%m-%Y %H:%M:%S'))

im = m2.imshow(result, origin='upper', cmap = palette, norm = pal_norm)
m2.colorbar(im,location='bottom')

savefig('%s.jpeg' % (ds_filename[-61:]), bbox_inches='tight', dpi=300, transparent=False) #transparent=True
!mv SOLAB*.jpeg /media/SOLabNFS2/tmp/
```
