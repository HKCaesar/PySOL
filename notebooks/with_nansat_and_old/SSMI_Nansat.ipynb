```python
# Helpers functions and classes
class Timer(object):
    """
    Timer — counting the elapsed spent.
    >>> with Timer("create ice boundary"):
    >>>     sime_func()...
    """
    def __init__(self, name=None):
        self.name = name

    def __enter__(self):
        self.tstart = time.time()

    def __exit__(self, type, value, traceback):
        if self.name:
            print '[%s]' % self.name,
        print 'Elapsed: %s seconds' % (time.time() - self.tstart)
        print "=" * 60
```

```python
class GranuleSSMI(object):
    """Class describe SSMI snapshot files.
    attributes: satellite name, created time, time range type"""
    
    processed = False
    
    def __init__(self, filename, path, root="/media/SOLabNFS/store/satellite/ssmi/", is_weekly=False):
        self.filename = filename
        self.path = path
        self.root = root
        self.created_at = None
        self.range_type = "day"
        self.is_weekly = is_weekly
        self.name_parser()
    
    def __repr__(self):
        return "<SSMI granule: %s %s>" % (self.range_type, self.filename)
        
    def name_parser(self):
        """Initializing next attributes: satellite, range_type, created_at"""
        m = re.match(ArcticIceApplication.pattern, self.filename)
        #try:
        self.satellite_number = m.group('ss')

        year = int(m.group('year'))
        month = int(m.group('month'))
        day = 1 if m.group('day') is None else int(m.group('day'))

        self.created_at = datetime.date(year, month, day)
        
        if self.is_weekly:
            self.range_type =  "week"
        elif m.group('day') is None and not m.group('date_type'):
            self.range_type = "month"
        elif not m.group('date_type'):
            self.range_type = "day" 
        elif m.group('date_type') == '_d3d':
            self.range_type = "3days"

        #except AttributeError:
        #    print "%s wrong" % self.filename
    
    def readFile(self):
        self.data = ArcticIceSSMIReader(os.path.join(self.path, self.filename))
        return self.data
        #print len(self.data)
        #print len(self.data['ice'])
        #print len(self.data['ice'][0])
        
    def createBorder(self):
        """ Create polygon and array of border's point for ice boundary"""
        
    #def createNansat(self):
    #    """ Create nansat for ... """
        
```

```python
def ArcticIceSSMIReader(filename, lonmin=0.125, lonmax=359.875, latmin=-89.875, latmax=89.875 ):
    """Reader for SSMI files"""
    xdim = 1440
    ydim = 720
    numvar = 4
        
    #scale = [.2, .3, .01, .1]  # scale factor for variables
    #offset = [0., 0., -0.05, 0.]
    
    data_final = {}.fromkeys(['lon', 'lat', 'ice'])
    
    #data_final['lon'] = np.linspace(-180.0, 180.0, xdim)
    #data_final['lat'] = np.linspace(-90.0, 90.0, ydim)
    #data_final['lat_grid'], data_final['lon_grid']= np.meshgrid(
    #    data_final['lat'], 
    #    data_final['lon'], 
    #    sparse=False, 
    #    indexing='ij'
    #)
    #lon = 0.25*xdim- 0.125
    #lat = 0.25*ydim-90.125
    
    data_final['lon']=np.zeros((xdim))
    data_final['lat']=np.zeros((ydim))
    
    for i in xrange(xdim):
        data_final['lon'][i] = 0.25 * (i + 1) - 0.125
       # print data_final['lon'][i]
    for y in xrange(ydim):
        data_final['lat'][y]=0.25*(y+1) - 90.125
       # print data_final['lat'][y]
    
    data_final['lon_grid'], data_final['lat_grid']= np.meshgrid(
        data_final['lon'],
        data_final['lat'],
    )

    f = gzip.open(os.path.join(filename), 'rb')
    buffer = f.read()
    f.close()
        
    data = np.frombuffer(buffer, dtype=np.uint8)\
        .reshape(numvar * ydim * xdim)
        
    map2 = np.reshape(data, (numvar, ydim, xdim))
    data_test = np.zeros((numvar, ydim, xdim))
    
    tmp = np.squeeze(map2[0,:,:])
    
    data_final['ice'] =  np.array(np.where(tmp != 252, 0, 1))
    data_final['land'] =  np.array(np.where(tmp != 255, 0, 0.5))

    # flip the lat_grid/land/ice so that North would be +90 at top
    data_final['lat_grid'] = data_final['lat_grid'][::-1]
    data_final['ice'] = data_final['ice'][::-1]
    data_final['land'] = data_final['land'][::-1]

    return data_final 
```

```python
class ArcticIceApplication(object):
    """Application class"""
    pattern = "(?P<ss>f\d{2})_" \
          "(?P<year>\d{4})" \
          "(?P<month>\d{2})" \
          "(?P<day>\d{2})?" \
          "(?P<version>(\w\d))" \
          "(?P<date_type>\w*).gz"

    def __init__(self, ssmi_root="/media/SOLabNFS/store/satellite/ssmi/"):
        self.ssmi_root = ssmi_root
        self.ssmi_granules = {
            'day': OOBTree.OOBTree(),
            '3days':OOBTree.OOBTree(),
            'week': OOBTree.OOBTree(),
            'month': OOBTree.OOBTree() }
    
    def registrationSSMI(self):
        """Search images in the root folder"""
        os.chdir(self.ssmi_root)
        count = 0
        with Timer("Scanning ssmi folder"):
            for (path, dirs, files) in os.walk(self.ssmi_root):
                is_weekly = True if path.find('weeks') != -1  else False
                for filename in files:
                    if re.match(ArcticIceApplication.pattern, filename):
                        granule = GranuleSSMI(filename, path, self.ssmi_root, is_weekly)
                        self.ssmi_granules[granule.range_type][filename] = granule
                        
                        count += 1
                        if count % 10000 == 0: 
                            print "- %d was counted" % count
        print "Total count of initialized snapshots: %d" % count
    
    def __del__(self):
        try: 
            self.connection.close()
        except AttributeError: 
            pass
```
