```python
def plot_plankton_2():
	from mpl_toolkits.basemap import Basemap
	import matplotlib.pyplot as plt
	plt.close('all')
	fig = figure(figsize=(10,10))
	
    # global ortho map centered on lon_0,lat_0
	lat_0=50.; lon_0=-55.
	m1 = Basemap(projection='ortho',lon_0=lon_0,lat_0=lat_0,resolution=None)
	ax = fig.add_axes([0.1,0.1,0.8,0.8],axisbg='k')
    # sliced ortho map
	m = Basemap(projection='ortho',lon_0=lon_0,lat_0=lat_0,resolution='l',\
				llcrnrx=0.,llcrnry=0.,urcrnrx=m1.urcrnrx/2.,urcrnry=m1.urcrnry/2.)
	cmap = plt.get_cmap('jet')
	cmap.set_bad('grey', 0)

	# reproject data
	px, py = m(x,y)
	for i, t in enumerate(ts):           
		ax.cla()
		dt = datetime.datetime.fromtimestamp(t)	
        print dt
        plankton = plankton_source[i][0][::-1,::]        
        plankton = np.ma.masked_where(isnan(plankton), plankton)  		
        plankton_ma = np.ma.array(plankton, mask=np.where((plankton>=0) & (plankton<=0), True, False))    
        cb = m.pcolor(px, py, plankton_ma, alpha=1.0, cmap=cmap)
        m.colorbar(cb)
        m.drawcoastlines(linewidth=0.25)
        m.drawcountries(linewidth=0.25)
        m.fillcontinents(color='coral',lake_color='blue')
        m.drawmapboundary(fill_color='blue')
        m.drawmeridians(np.arange(0,360,30))
        m.drawparallels(np.arange(-90,90,30))   
        ax.set_title('Phytoplankton Concentration (as Carbon), mmol/m**3, %s' % dt.strftime('%d-%m-%Y'))    
        plt.savefig('data/plankton_2/%s.jpg' % dt.strftime('%Y%m%d'))                    
```
