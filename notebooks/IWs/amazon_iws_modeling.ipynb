### Some theory

Дисперсионное соотношение для длинных (мелкая вода) гравитационных волн:

$$ \omega = \sqrt{gh} K. $$

Для внутренних волн в двухслойной жидкости:

\begin{equation}\label{eq:eq1}
    \omega = \sqrt{ g \frac{ \Delta \rho} {\rho} h } K,
\end{equation}

С учётом течения:

\begin{equation}\label{eq:eq2}
    (\omega - KU )^2 = g \frac{ \Delta \rho} {\rho} h K^2,
\end{equation}

где в выбранной системе координат $K = \sqrt{k_x^2 + k_y^2}$.

Соотношение, связывающее пространственную изменчивость частоты с пространственной неоднородностью (дивергенцией) скорости течения и пространственной неоднородностью собственной фазовой скорости, обусловленной неоднородностью перепада плотности и толщины верхнего слоя. Это соотношение имеет вид:

Пакет ВВ распространяется от берега и входит в область геострафического течения направленного к берегу. Наблюдается рефракция ВВ (преломление фронта волны). Пусть глубина залегания пикноклина $h$ не меняется вдоль течения и остаётся постоянной, меняясь только в направлении перпендикулярном течению, вдоль оси $Х$ на Рисунке. Координатная система выбрана таким образом, что ось $X$ направлена поперёк течения, а ось $Y$ - вдоль течения.

$$ \omega = \sqrt {g \frac{ \Delta \rho} {\rho} h } \sqrt{ k_x^2 + k_y^2 } + k_xU_x + k_yU_y. $$

С учётом, что течение геострофично, $U_x=0$, $U_y = \frac{ g } {\rho f} \frac{\Delta p}{\Delta x} $, и потенциально, $ \frac{\partial k_y}{\partial x} - \frac{\partial k_x}{\partial y} = 0 $, $k_y=const$, а также уравнения гидростатического баланса, $ \frac{\partial p}{\partial z} = -g\rho $, получаем:

$$ 2\pi/T = \omega = \sqrt {g \frac{ \Delta \rho} {\rho} h(x) } \sqrt{ k_x^2 + k_y^2 } + k_yU_y = const = g \frac{ \Delta \rho} {\rho} h_0 K_0. $$

$$ \sqrt {g \frac{ \Delta \rho} {\rho} h(x) } \sqrt{ k_x^2 + k_y^2 } = g \frac{ \Delta \rho} {\rho} h_0 K_0 - k_yU_y. $$

$$ g \frac{ \Delta \rho} {\rho} h(x) \left( k_x^2 + k_y^2 \right) = \Bigl( C - k_y \frac{g}{f} \frac{\Delta \rho} {\rho} \frac{\partial h}{\partial x} \Bigr)^2. $$

$$ k_x^2 = \frac { \Bigl( C - k_y \frac{g}{f} \frac{\Delta \rho} {\rho} \frac{\partial h}{\partial x} \Bigr)^2 }
{ g \frac{ \Delta \rho} {\rho} h(x) } - k_y^2, $$

где $ C = g \frac{\Delta \rho} {\rho} h_0 K_0. $

Пусть глубина залегания пикноклина $h(x)$ задана гиперболическим тангенсом, $h(x)=tanh(x)$.

В общем случае плавучесть $ \frac{ \Delta \rho} {\rho}$ зависит от температуры и солёности, для простоты будем использовать уравнение состояния в приближении Буссинеска, с линейной зависимотью только от температуры:
$$ \frac{ \Delta \rho} {\rho} = \alpha (T-T_0) = \alpha \Delta T,$$
где $\alpha=-2\cdot10^{-4} [^\circ C^{-1}]$, коэффициент термического расширения (coefficient of thermal expansion), а $T_0$ - референсная постоянная температура.
Пусть перепад температур составляет десять градусов, $\Delta T=10$, тогда $ \frac{ \Delta \rho} {\rho} = -2\cdot10^{-3}$.

Предположим, что изначально пакет ВВ распространяется из области где нет течения \ref{eq:eq1} и, попадая в область с течением \ref{eq:eq2}, падает на течение под углом $\phi_0$, тогда $\tan \phi_0 = \frac{k_{0y}} {k_{0x}} $. Принимая во внимание, что $K^2 = k_x^2 + k_y^2$, выразив $k_x$ через $k_y$ и $\tan \phi$, можно записать:

$$ K^2 = \frac{k_y^2}{\tan^2 \phi} + k_y^2, $$

откуда:

$$ k_y^2 = k_{0y}^2 = \frac { \tan^2 \phi_0 } { 1 + \tan^2 \phi_0 } \cdot K_0^2 = const, $$

где $K_0$ оценим по РСА изображению, зная период приливной волны, генерирующей ВВ.

```python
from numpy import tanh, cosh, tan, arctan2, linspace, gradient
%pylab inline
import matplotlib.pyplot as plt
```

```python
# predefining constants
lat = 4
# At a latitude of lat, the acceleration due to gravity at sea level is approximately
g = 9.780327 * ( 1 + .0053024*sin(lat*pi/180)**2 - .0000058*sin(2*lat*pi/180)**2 )
f = 2*7.3E-5*sin(lat*pi/180)  # sin(-lat) minus for the southern hemisphere
D = -2e-3

phi0 = (45)*pi/180

h0 = -50
K0 = 2*pi/(12.4*3600*2) # 2pi/(T*Cf)
ky = tan(phi0)/sqrt(1+tan(phi0)**2) * K0
C = g*D*h0*K0

print g, '\n', f, '\n', D, '\n', ky, '\n', K0, '\n', C
```

```python
L  = 50e3   # Current half-width L in [m]
Z  = -1.5e2 # Pychnoclyne depth
np = 1e1    # number of points
X = linspace(0,2*L,np)
H = linspace(h0,Z,np)
# Hx = gradient(H,(2*L)/np)
Hx = gradient(H, (2*L+Z*sign(Z)+h0*sign(h0))/np)

# Найдём компоненту волнового вектора kx
kx2 = (C - ky * g/f*D*Hx)**2 / (g*D*H)**2 - ky**2
# kx2 = (C - ky * 2       )**2 / (g*D*H)**2 - ky**2
phi = arctan2(ky**2,kx2)*180/pi
```
