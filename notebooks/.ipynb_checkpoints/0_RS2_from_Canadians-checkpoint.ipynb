## Polarization Ratio coeff for the bragg waves

### Some theory

$$ p_{B} = \frac{ \Bigl| G_h^2 \Bigr| \left( 1 + g_h s_i^2 \right) } { \Bigl| G_v^2 \Bigr| \left( 1 + g_v s_i^2 \right) } $$

as in GRSL paper: </br>
`V. N. Kudryavtsev, B. Chapron, A. G. Myasoedov, F. Collard, and J. A. Johannessen, “On Dual Co-Polarized SAR Measurements of the Ocean Surface,” IEEE Geoscience and Remote Sensing Letters, vol. 10, no. 4, pp. 761–765, Jul. 2013.`

How did we achieve this formula?</br>
Using RIM part1 formulae: </br>
`V. N. Kudryavtsev, D. Hauser, G. Caudal, and B. Chapron, “A semiempirical model of the normalized radar cross-section of the sea surface 1. Background model,” Journal of Geophysical Research, vol. 108, no. C3, p. 24, Jan. 2003.`</br>

$$ \sigma_{br}^p = \sigma_{0br}^p \Bigl( 1 - M_{t0}^p \zeta_x + \frac{ \tilde{B_r} }{ B_{r0} } + g^p \zeta_x^2 - \frac{ M_{t0}^p }{ B_{r0} } \zeta_x \tilde{B_r} \Bigr) $$

Hereinafter $p$ indicates $hh$ or $vv$.

When averaging Taylor series decomposition</br>
$$ f(a)+\frac {f'(a)}{1!} (x-a)+ \frac{f''(a)}{2!} (x-a)^2+\frac{f^{(3)}(a)}{3!}(x-a)^3+ \cdots $$

Mean of the oscilating function turns to zero:</br>
$$ \overline{x-a} \to 0 $$

Let us check this.</br>
The mean of a function $f(x)$ over the interval $(a,b)$ is defined by</br>

$$ \bar{f}=\frac{1}{b-a}\int_a^bf(x)\,dx $$

Assume</br>

$$ f(t) = A \cos{\omega t} $$

$$a = -T/2$$
$$b = +T/2$$

Then</br>

$$ \overline{f(t)} = \frac{1}{T}\int_{-T/2}^{T/2}f(t)\,dt $$


Calculate the improper integral</br>

$$ \overline{f(t)} = \lim_{T\to\infty} \frac{1}{T}\int_{-T/2}^{T/2} A \cos{\omega t} \,dt = \lim_{T\to\infty} \frac{A \sin{\omega t}}{\omega T} \Bigr|_{-T/2}^{+T/2} = \lim_{T\to\infty} \left( \frac{ A \sin{(\omega T/2)}}{ \omega T } - \frac{ A \sin{(-\omega T/2)} }{ \omega T } \right) =  2A \lim_{T\to\infty} \left( \frac{ \sin{(\omega T/2)} }{ \omega T } \right) \to 0$$

```python
plt.close('all')
plt.plot(sin(arange(1,1e2,0.1))/arange(1,1e2,0.1))
```

$ \tilde{B}_r $ - is the spectrum variation due to short waves (SW) modulatoins by tilting waves.

Same applies to $ \tilde{B_r} $, as it oscilates as well

Following the above, when considering the SAR pixel we average over many bragg waves, which leads to:</br>
$$ \sigma_{br}^p = \sigma_{0br}^p \left( 1  + g^p \zeta_x^2 \right) $$

$g^p$ - is a polarization coefficient accounting for the effect of the tilting waves in the second order given by:</br>

$$ g^v = \left( \frac {1}  {2 \sigma_{0br}^v} \right) \partial{}^2 \sigma_{0br}^v / \partial{}^2 \theta $$
for the VV polarized scattering

$$ g^h = \left( \frac {1}  {2 \sigma_{0br}^h} \right) \partial{}^2 \sigma_{0br}^h / \partial{}^2 \theta{}^2 + \left( 2 / \sin^2{\theta_0} \right) \left| G_{v0} \right| / \left| G_{h0} \right| \left( \zeta_y^2 / \zeta_x^2 \right) $$
for the HH polarized scattering


For the Sea surface Bragg scattering geometric coefficients are:

$$ \left| G_{v0} \right|^2 = \frac{ \cos^4{\theta} \left( 1 + \sin^2{\theta} \right) ^2 } { \left( \cos \theta + 0.111 \right) ^4 } $$

$$ \left| G_{h0} \right|^2 = \frac{ \cos^4{\theta} } { \left( 0.111 \cos \theta + 1 \right) ^4 } $$

Assume the spectrum $B$ is a constant 1, as we don't need to calculate the real NRCS, but just want to see the contribution of each term.</br>
Also one may notice that in PR the spectrum $B$ will be reduced as it exists in both the nominator and denominator.</br>
If we would like to calculate the real NRCS we would need to describe the real model, as in RIM for example.

### Calculate $\zeta_y^2$  and $\zeta_x^2$

How do we calculate $\zeta_y^2$  and $\zeta_x^2$ - the slopes of the tilting waves cross and along the incidence plane?

Follwong Cox and Munk:</br>
`Cox, C., & Munk, W. (1954). Measurement of the Roughness of the Sea Surface from Photographs of the Sun’s Glitter. Journal of the Optical Society of America, 44(11), 838–850. doi:10.1364/JOSA.44.000838`</br>
and Phillips:</br>
`Phillips, O. M. (1977). The dynamics of the upper ocean. (G. K. Batchelor & J. W. Miles, Eds.) (Vol. 88, p. 344). Cambridge University Press.`

Suppose that the surface slope distribution is isotropic, which leads to $\zeta_x^2 = \zeta_y^2$, then

$$ \zeta_x^2 = \zeta_y^2 = \frac{1}{2} \alpha \log{(k_s/k_0)} $$

here $k_0 = g/U^2 $ - lower limit of the saturation range,</br>
$k_s$ - upper limit and following RIM papers it is 4 times less than $k_r$: $k_s = 1/4 k_r$,</br>
Cox and Munk experiments give $\alpha=4.6 \cdot 10^{-3}$

```python
# When using diff do not set steps smaller than 1, there is no option to set the sample distances in each direction in diff as in gradient
# As the theta function is smooth we can use step 1 an then interpolate the results
step = (IncidenceAngle.max() - IncidenceAngle.min()) / IncidenceAngle.size
#step = 1
```

```python
theta_min = 20.0
theta_max = 60.0
```

```python
theta = arange(theta_min,theta_max,step)*pi/180
g = 9.81
U = 7
k_r = 2*pi/5.6e-02
zeta_x2 = .5*4.6e-3*log10(.25*k_r/g*U**2)
zeta_y2 = zeta_x2
```

One may note that the wind speed here was chosen to be 7m/s as the mean wind over Ocean.</br>
$k_r$ corresponds to C-band wavelength

### Calculate scattering $G_p$ and polarization $g^p$ coefficients

```python
theta = arange(theta_min-10,theta_max+30,step)*pi/180
step = step*pi/180
B = 1
k_B = 2*k_r*sin(theta)
W = k_B**(-4)*B
Gv0 = (cos(theta)**4 * (1 + sin(theta)**2 )**2 / ( cos(theta) + 0.111 )**4)
Gh0 = (cos(theta)**4 / ( 0.111*cos(theta) + 1 )**4)
sigma_v_br = 16*pi*k_r**4*Gv0*W
sigma_h_br = 16*pi*k_r**4*Gh0*W
```

We will use the gradient as it is computed using central differences in the interior and first differences at the boundaries.</br>
Using diff will lead to incorrect values on the edges, but the gradient as well, as wil be shown further on the plot.</br>
That is why we will find the gradient on a larger angles range (from 10 to 70 degrees) and then take only the 20 to 60 values.

```python
sigma_v_br = append(sigma_v_br[0], sigma_v_br)
sigma_v_br = append(sigma_v_br, sigma_v_br[-1])

sigma_h_br = append(sigma_h_br[0], sigma_h_br)
sigma_h_br = append(sigma_h_br, sigma_h_br[-1])
```

```python
gv2 = .5/sigma_v_br[1:-1]*diff(sigma_v_br, n=2)/step**3
gv  = .5/sigma_v_br[1:-1]*gradient(gradient(sigma_v_br[1:-1], step), step)
gh  = .5/sigma_h_br[1:-1]*gradient(gradient(sigma_h_br[1:-1], step), step)
gh_ = .5/sigma_h_br[1:-1]*gradient(gradient(sigma_h_br[1:-1], step), step) + 2 / sin(theta)**2 * Gv0/Gh0
```

```python
plt.close('all')
plt.semilogy(theta*180/pi, gv2, 'r', label='$gv$ diff')
plt.hold('on')
plt.semilogy(theta*180/pi, gv, 'b', label='$gv$ gradient')
plt.semilogy(theta*180/pi, gh, 'k', label='$gh$ w/o $Gv0/Gh0$')
plt.semilogy(theta*180/pi, gh_, 'k:', label='$gh$ with $Gv0/Gh0$')
plt.semilogy((20,20), (1e-4,1e6), 'g')
plt.semilogy((60,60), (1e-4,1e6), 'g')
#plt.axis((10., 70., 1e-1, 1e3))
plt.legend()
```

```python
plt.close('all')
plt.plot(theta*180/pi, gv2, 'r', label='$gv$ diff')
plt.hold('on')
plt.plot(theta*180/pi, gv, 'b', label='$gv$ gradient')
plt.plot(theta*180/pi, gh, 'k', label='$gh$ w/o $Gv0/Gh0$')
plt.plot(theta*180/pi, gh_, 'k:', label='$gh$ with $Gv0/Gh0$')
plt.axis((10., 70., 0, 100))
plt.legend()
```

Taking only the range from `theta_min` to `theta_max`. In our case 20 to 60 degrees

```python
theta = rad2deg(theta)
gv = gv[((theta>=theta_min) & (theta<=theta_max))]
gh = gh[((theta>=theta_min) & (theta<=theta_max))]
gh_ = gh_[((theta>=theta_min) & (theta<=theta_max))]
Gh0 = Gh0[((theta>=theta_min) & (theta<=theta_max))]
Gv0 = Gv0[((theta>=theta_min) & (theta<=theta_max))]
theta = theta[((theta>=theta_min) & (theta<=theta_max))]
```

```python
plt.close('all')
plt.semilogy(theta, gv, 'b', label='$gv$')
plt.semilogy(theta, gh, 'k', label='$gh$ w/o $Gv0/Gh0$')
plt.semilogy(theta, gh_, 'k:', label='$gh$ with $Gv0/Gh0$')
plt.legend()
```

```python
plt.close('all')
plt.plot(theta, gv, 'b', label='$gv$')
plt.plot(theta, gh, 'k', label='$gh$ w/o $Gv0/Gh0$')
plt.plot(theta, gh_, 'k:', label='$gh$ with $Gv0/Gh0$')
plt.axis((20., 60., 0, 1e2))
plt.legend()
```

Now the plot looks much better and more than that - closer to truth

## Calculate PR

```python
PRmodel  = Gh0*(1+gh *zeta_x2) / ( Gv0*(1+gv*zeta_x2) )
PRmodel_ = Gh0*(1+gh_*zeta_x2) / ( Gv0*(1+gv*zeta_x2) )
PRbr     = Gh0 / Gv0
```

Note that here we plot each 1e3 element to achieve nice plot, as there are too many elements for dotted plot

```python
plt.close('all')
plt.figure(figsize=(13,7))
plt.plot(theta[::1e3], PRmodel [::1e3], 'b-', label='$gh$ w/o $Gv0/Gh0$')
plt.plot(theta[::1e3], PRmodel_[::1e3], 'r.', label='$gh$ with $Gv0/Gh0$')
plt.plot(theta[::1e3], PRbr[::1e3], 'k-', label='pure Bragg')
plt.legend()
```

## Non Polarized part due to wave breaking

$$\sigma_{wb} = \sigma_0^{vv} - \Delta\sigma / \left( 1- p_B \right)$$

```python
# As the IncidenceAngle function is smooth we can use step 1 an then interpolate the results.
# If you have set the step from IncidenceAngle use commented PRmodel line below
PRmodel = PRmodel[((theta>=IncidenceAngle.min()) & (theta<=IncidenceAngle.max()))]
PRmodel = resize(PRmodel, SigmaHHwnr.shape)
# This line if you have used step = 1
#PRmodel = tile(PRmodel, (line[-1,0]+1, 1))
#PRmodel = resize(PRmodel, SigmaHHwnr.shape)
PRmodel.shape
```

```python
SigmaWb = SigmaVVwnr - PD/(1 - PRmodel)
```

<script>

$(function(){

  function list_el(h){
    href = $(h).find('a.anchor-link').attr('href')
    text = $(h).find('a.heading-anchor').text()
    return "<a href='" + href + "'>" + text + "</a>"
  }

  function lvl(header){
    if (header !== undefined)
      return $(header).prop('tagName').toLowerCase().substring(1);
    else return 0;
  }

  window.toc = function(){
    headers = $('h1, h2, h3, h4');
    html = "<ol class='nested'>";
  
    $.each(headers , function(i, h) {
      if (lvl(h) == lvl(headers[i+1])) {
        html += "<li class='nested'>" + list_el(h) + "</li>" ;
      }
      else if ( lvl(h) < lvl(headers[i+1]) ) {
        html += "<li class='nested'>" + list_el(h) + "<ol class='nested'>";
      }
      else if (lvl(h) > lvl(headers[i+1]) ) {
        step = lvl(h) - lvl(headers[i+1])
        html += "<li class='nested'>"+ list_el(h) +"</li></ol></li>";
        if (step > 1){
            html+="</ol></li>"
        }
        
      }
    }) 
    html += "</ol>"
  
    $('#toc').html(html)
    $('#toc-static').html(html)
    
  }
  toc()
  
  $('#toc-wrapper .hide-btn').click(function(){
      $('#toc').slideToggle()
      
      $('#toc-wrapper').toggleClass('closed')
      
      if ($('#toc-wrapper').hasClass('closed')){
        $('#toc-wrapper .hide-btn').html('[show]')
      } else {
          $('#toc-wrapper .hide-btn').html('[hide]')
      }
      
      return false
  })
  $('#toc-update').click(function(){toc(); return false;})
  $(window).resize(function(){
      max_height = $(window).height() - 300;
      $('#toc').css({maxHeight: max_height});
  })

})
</script>
