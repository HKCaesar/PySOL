# Testing gradient function

```python
v = arange(-2.0,2.0,0.2)
[x,y] = meshgrid(v,v)
```

```python
close('all')
imshow(x)
colorbar()
```

```python
z = x * exp(-x**2 - y**2)
[py,px] = gradient(z,.2)
```

```python
plt.close('all')
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.imshow(px)
plt.colorbar()
plt.subplot(122)
plt.imshow(py)
plt.colorbar()
```

```python
x = arange(0,100,0.1)
y = x**2
```

```python
py = gradient(y, 0.1)
pyy = gradient(py, 0.1)
```
