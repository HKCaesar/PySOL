```python
from pylab import *
import pyresample as pr
# OPeNDAP client: http://pydap.org
from pydap.client import open_url
import numpy.ma as ma
import matplotlib.colors as colors
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import os, errno
import time
from os import walk
import re
import gc
import argparse
```

```python
projections = [{'name': 'epsg_4326', 'inch_size': (14.40, 7.20)},
               {'name': 'epsg_3413', 'inch_size': (14.40, 14.40)}]

projection = projections[0]


url = 'http://opendap.solab.rshu.ru:8080/opendap/hyrax/allData/OISST-AVHRR-AMSR-V2/2004/AVHRR-AMSR/amsr-avhrr-v2.20040104.nc.gz'
clf()
dataset = open_url(url)
pp_name = 'sst'
pp_value = dataset[pp_name][:][pp_name][:]
pp_value = resize(pp_value, (pp_value.shape[2], pp_value.shape[3]))

scale_factor = dataset[pp_name].attributes['scale_factor']
add_offset = dataset[pp_name].attributes['add_offset']
fill_value = dataset[pp_name].attributes['_FillValue']

valid_min = dataset[pp_name].attributes['valid_min']
valid_max = dataset[pp_name].attributes['valid_max']

pp_value = ma.masked_where(pp_value > valid_max, pp_value)
pp_value = ma.masked_where(pp_value < valid_min, pp_value)
pp_value = np.float64(ma.masked_values(ma.masked_equal(pp_value, None, copy=False), fill_value, copy=False))\
                 * scale_factor + add_offset
    
vmin = valid_min * scale_factor + add_offset
vmax = valid_max * scale_factor + add_offset

# data_array = np.rot90(sst)
# data_array = np.rot90(data_array)

# pp_value = ma.masked_where(pp_value == 0, pp_value)
data_array = pp_value

# clf()
# imshow(data_array)
lat = dataset.lat[:]
lon = dataset.lon[:]
lon = ma.where(lon <= 180, lon, lon - 360).data

# swath_def = pr.geometry.SwathDefinition(lons=lon, lats=lat)
x, y = np.meshgrid(lon, lat)
grid_def = pr.geometry.GridDefinition(x, y)




area_def = pr.utils.load_area('/media/SOLabNFS/http/public/daniel/areas.cfg', '%s_crude'%(projection['name']))

result = pr.kd_tree.resample_nearest(grid_def, data_array.ravel(), area_def, radius_of_influence=30000,
                                     fill_value=fill_value, nprocs=2)

try:
    plt.close('all')
except AttributeError as e:
#     print e
    pass

#specify figure size in inches:
fig = figure(num = 1, figsize=(projection['inch_size']), facecolor=None, edgecolor=None, frameon=False)


ax = axes([0.,0.,1.,1.])
ax.set_axis_off()
fig.add_axes(ax)
m = pr.plot.area_def2basemap(area_def, resolution=None)

palette = cm.jet
# palette.set_bad('grey', 0)
pal_norm = colors.Normalize(vmin=vmin, vmax=vmax, clip = False)

#m.drawcoastlines(linewidth=0.5, color='black', antialiased=1, ax=None, zorder=None)
im = m.imshow(result, origin='upper', cmap = palette, norm = pal_norm)

savefig('/tmp/1.png', dpi=100, transparent=True, pad_inches=0)
```
