### Importing different modules

What is a Python Module and what is its difference from Python Package?
A Module is a single file (or files) consisting of Python code. A module can define functions, classes, as well as variables. A module can also include runnable code.
A Package is a collection of Modules in directories that give a package hierarchy.
Packages are a way of structuring Python's module namespace by using "dotted module names".
See Python official tutorial for more details:
http://docs.python.org/2/tutorial/modules.html

In our case all the modules we import from the nansat form  the Nansat package.

## Part 2. Start playing with the data

### Reading the data

```python
n = Nansat(iPath + fileName, mapperName='asar')
ncep = Nansat(ncepPath + ncepFileName, mapperName='ncep')
```

### Reprojecting NCEP onto ASAR

```python
# Reproject NCEP onto ASAR
ncep.reproject(n, eResampleAlg=3)
```

### Set the variables for further Wind calculations

The sigma nought image is derived from ESA’s ASAR level 1 WSM products following:</br>
Rosich B., Meadows P., Absolute calibration of ASAR Level 1 products, ESA/ESRIN, ENVI-CLVL-EOPG-TN-03-0010, Issue 1, Revision 5, October 2004</br>
https://earth.esa.int/web/guest/missions/esa-operational-eo-missions/envisat/content?p_r_p_564233524_assetIdentifier=absolute-calibration-of-asar-level-1-products-generated-with-pf-asar-4503

```python
#raw_counts = n[1]
inc_angle = n[2]
sigma0 = n[3]
u = ncep[1]
v = ncep[2]
w = ncep[3]
windDir = ncep[4]
```

### Filtering the SAR data before processing

Using Wiener filter from scipy.signal module to filter out the speckle noise.</br>
Wiener and Frost filters should be used for beginner, while</br>
Lee and Median filters should be used for more experienced physicists.

Import time module to estimate elapsed calculation time

```python
from scipy.signal import wiener
from time import time
```

```python
currtime = time()
sigma0w = wiener(sigma0, mysize=(7,7), noise=None)
print 'Wiener filter elapsed time: %f', ( time() - currtime )
```

Using Cython Lee filter implementation [https://github.com/josephmeiring/LeeFilter](https://github.com/josephmeiring/LeeFilter).</br>
Cython [http://www.cython.org/](http://www.cython.org/) is an optimising static compiler for both the Python programming language and the extended Cython programming language (based on Pyrex).</br>
With pyximport Python modules are compiled to .pyc files while the compilation process is hidden from the end-user.</br>
For instance in our case we have a Cython module called lee.pyx, with Pyximport you can import it in a regular Python module like this:

```python
import pyximport; pyximport.install(build_in_temp=False)
from lee import lee
```

```python
currtime = time()
sigma0l = lee(sigma0.astype(double))
print 'Lee filter elapsed time:', ( time() - currtime )
```

We set the scale to 20 picking every 20th pixel from the data array to reduce the image size we are going to further process

```python
scale = 20
sigma0wCrop = 10*log10(sigma0w[::scale,::scale])
sigma0lCrop = 10*log10(sigma0l[::scale,::scale])
```

Plot to compare Wiener and Lee filters results

```python
plt.close()
plt.figure(figsize=(15,8))
plt.subplot(121)
plt.imshow(sigma0wCrop, vmin=-20, vmax=0)
plt.gray()
plt.colorbar()
plt.subplot(122)
plt.imshow(sigma0lCrop, vmin=-20, vmax=0)
plt.gray()
plt.colorbar()
```

### Calulating Wind Speed from SAR in parallel (on several cores)

```python
#NB! WINDDIR = 0 WHEN WIND BLOWS TOWARDS RADAR!
wDrel = 90 + n.upwards_azimuth_direction() - windDir[::scale,::scale] 
i = inc_angle[::scale,::scale]
try:
    wPar = rcs2windPar(sigma0wCrop, cmdv=4, windir=wDrel, theta=i, nprocs=10)
except Exception:
    wPar = rcs2wind(sigma0wCrop, cmdv=4, windir=w, theta=i)
print "Testing CMOD4 passed, Wind =", wPar.mean()
```

### Creating "Nice" SAR Image ( which is the Roughness)

```python
#~ NICE image (roughness)
pol = n.bands()[3]['polarization']
if pol == 'HH':
    ph = (2.20495, -14.3561e-2, 11.28e-4)
    sigma0_hh_ref = exp( ( ph[0]+inc_angle*ph[1]+inc_angle**2*ph[2])*log(10) )
    roughness = sigma0w/sigma0_hh_ref
elif pol == 'VV':
    pv = (2.29373, -15.393e-2, 15.1762e-4)
    sigma0_vv_ref = exp( ( pv[0]+inc_angle*pv[1]+inc_angle**2*pv[2])*log(10) )
    roughness = sigma0w/sigma0_vv_ref
```

### Creating new bands for saving

```python
#~ Create new band
#GDALDataType { 
#  GDT_Unknown = 0, GDT_Byte = 1, GDT_UInt16 = 2, GDT_Int16 = 3, 
#  GDT_UInt32 = 4, GDT_Int32 = 5, GDT_Float32 = 6, GDT_Float64 = 7, 
#  GDT_CInt16 = 8, GDT_CInt32 = 9, GDT_CFloat32 = 10, GDT_CFloat64 = 11, 
#  GDT_TypeCount = 12 
#}
n.add_band(bandID=4, array=roughness, \
   parameters={'name':'roughness', \
   'wkv': 'surface_roughness', \
   'dataType': 7, 'minmax': '0 2'})

n.add_band(bandID=5, array=sigma0w, \
   parameters={'name':'sigma0', \
   'wkv': 'surface_backwards_scattering_coefficient_of_radar_wave', \
   'dataType': 7, 'minmax': '-20 0'})

n.add_band(bandID=6, array=wPar, \
   parameters={'name':'wind_speed', \
   'wkv': 'surface_wind_speed_reconstructed_with_CMOD4', \
   'dataType': 7, 'minmax': '0 20'})

n.add_band(bandID=7, array=w, \
   parameters={'name':'wind_speed', \
   'wkv': 'surface_wind_speed_NCEP_GFS', \
   'dataType': 7, 'minmax': '0 20'})
```

### Cropping the Baltic Sea region

```python
# Cropping the baltic sea region
latlim = array((53, 65))
lonlim = array((12, 33))
```

### Reproject image into Lat/Lon WGS84 (Simple Cylindrical) projection

Google Earth uses Simple Cylindrical projection with a WGS84 datum for its imagery base.</br>
You can find more info in Google Earth User Guide:</br>
http://laearth.ohsep.louisiana.gov/intl/en/client/ug_importdata.html#note

```python
# Reproject image into Lat/Lon WGS84 (Simple Cylindrical) projection
# 1. Cancel previous reprojection
# 2. Get corners of the image and the pixel resolution
# 3. Create Domain with projection
# 4. Reproject
# 5. Write image
n.reproject() # 1.
lons, lats = n.get_corners() # 2.
# Pixel resolution
pxlRes = array(distancelib.getPixelResolution(array(lats), array(lons), n.shape(), 'deg'))
pxlRes = pxlRes*7 # make the resolution worser
srsString = "+proj=latlong +datum=WGS84 +ellps=WGS84 +no_defs"
extentString = '-lle %f %f %f %f -tr %f %f' % (min(lonlim), min(latlim), \
                max(lonlim), max(latlim), pxlRes[1], pxlRes[0])
d = Domain(srs=srsString, ext=extentString) # 3.
n.reproject(d) # 4.
```

### Apply Land Mask

```python
# write one band to GeoTIFF for geoserver using write_figure method
wm = n.watermask(mod44path='/media/SOLabNFS/store/auxdata/coastline/mod44w/')
wmArray = wm[1]
```

### Finally export to Google Earth

```python
n.write_figure(fileName=oPath+fileName+'_roughness.png', bands=[4], \
                 mask_array=wmArray, mask_lut={0: [0,0,0]},
                 clim=[0,2], cmapName='gray', transparency=[0,0,0]) # 5.

# make KML file for the exported image
n.write_kml_image(kmlFileName=oPath+fileName+'_roughness.kml', kmlFigureName=oPath+fileName+'_roughness.png')
```
