```python
import time
def hsince2ymd(y,m,d,offset_day):
    """Returns year,month,day string for an offset day"""
    time_ymd=time.mktime((y, m, d, 1, 0, 0, 0, 0, 0))
    ltime=time.localtime(time_ymd+offset_day*60*60*24)
    return [str(ltime[0]),str( '%(#)02d' % {'#':(int(ltime[1]))}),str( '%(#)02d' % {'#':(int(ltime[2]))})]
```

```python
hsince2ymd(2010,1,1,dataset.time[-74]/24) # Determine latest date
```

```python
lats=dataset['latitude'].array[:,:]
lons=dataset['longitude'].array[:,:]
sit=array(dataset['sea_ice_thickness'][-1,:,:])
```

```python
# Pixel resolution
import distancelib

# we use pxlResWind for further pyresample radius_of_influence and sigmas
pxlResSMOS = asarray(distancelib.getPixelResolution(lats, lons, sit[0].shape, 'km'))

# Note pxlResWind is in KM, multiply by 1e3 for meters
print "SMOS cell resolution, %s km"  % pxlResSMOS
```
