```python
# Helper functions and classes
class Timer(object):
    """
    Timer — counting the elapsed spent.
    >>> with Timer("create ice boundary"):
    >>>     sime_func()...
    """
    def __init__(self, name=None):
        self.name = name

    def __enter__(self):
        self.tstart = time.time()

    def __exit__(self, type, value, traceback):
        if self.name:
            print '[%s]' % self.name,
        print 'Elapsed: %s seconds' % (time.time() - self.tstart)
```

## Интерфейс доступа к SSMI гранулам

```python
def get_data_from_ssmi_granule(filename, lonmin=0, lonmax=360, latmin=-90, latmax=90, filter_step=4, init_land=False):
    """Reader for SSMI files
    Parameters
    ----------
    filename : str
        This str should give absolute path and filename of granule
    filter_step: int 
        Horizontal step for sift data
    init_land: boolean
        True: add land mask to the data

    Returns
    -------
    data : dict
        data consist of lat_grid, lon_grid and ice mask
    error : float
        The error for the granule
    
    Examples
    ------
    >>> get_data_from_ssmi_granule('/media/SOLabNFS/store/satellite/ssmi/f10/bmaps_v07/y1994/m07/f10_19940709v7_d3d.gz', latmin=60.0, init_land=True)
    ... plt.imshow(data['ice'] + data['land'] * 5)
    """
    STEP = .25
    numvar = 4
    xdim, ydim = 1440, 720
    istep = int(1/STEP)
    
    data = {}.fromkeys(['lon', 'lat', 'ice', 'lat_grid', 'lon_grid'])
    
    data['lon'] = np.array([0.25 * (x + 1) - 0.125 for x in xrange(xdim)])
    data['lat'] = np.array([0.25 * (y + 1) - 90.125 for y in xrange(ydim)])
    
    data['lat_grid'], data['lon_grid']= np.meshgrid(data['lat'], data['lon'],
                                                    sparse=False, indexing='ij')
    
    # read binary ssmi file to temporary buffer in RAM
    f = gzip.open(os.path.join(filename), 'rb')
    temporary_buffer = f.read()
    f.close()
        
    data_buffer = np.frombuffer(temporary_buffer, dtype=np.uint8).reshape(numvar * ydim * xdim)
    map2 = np.reshape(data_buffer, (numvar, ydim, xdim))
    tmp = np.squeeze(map2[0,:,:])
    
    del(data_buffer)
    del(temporary_buffer)
    
    data['lon'] = data['lon'][int(lonmin * istep) : int(lonmax * istep):filter_step]
    data['lat'] = data['lat'][int(latmin * istep) + 360 : int(latmax * istep) + 360]
    data['ice'] = np.array(np.where(tmp != 252, 0, 1))
    
    data['unknown'] = np.array(np.where((tmp != 254) , 0, 1))
    data['bad'] = np.array(np.where((tmp != 253) , 0, 1))
    
    filtering_array = ['lat_grid', 'lon_grid', 'ice', 'unknown', 'bad']
    turn_array = ['ice', 'unknown', 'bad']
    
    if init_land:
        data['land'] =  np.array(np.where(tmp != 255, 0, 1))
        filtering_array.append('land')
        turn_array.append('land')
    
    # sift data
    for key in filtering_array:
        data[key] = data[key][int(latmin * istep) + 360: int(latmax * istep) + 360,\
             int(lonmin * istep): int(lonmax * istep): filter_step]
    
    # turn data
    for key in turn_array:
        data[key] = data[key][::-1]
        
    # calculate "error" for granule
    
    error = data['unknown'].sum() / (xdim *  ydim / float(filter_step))
            
    # fill black circle in NP
    for i in range(16):
        data['ice'][i] |=  data['unknown'][i] |  data['bad'][i] #[1 for _ in range(xdim/filter_step)]
    
    del(data['bad'])
    del(data['unknown'])
   
    return data, error
```

## Маска земли для изображений 

```python
def get_land_mask(lonmin=0, lonmax=360, latmin=-90, latmax=90, filter_step=4, 
    filename="/media/SOLabNFS/store/satellite/ssmi/f10/bmaps_v07/y1994/m07/f10_19940709v7_d3d.gz"):
    """Return land mask
    Parameters
    ----------
    filename : str
        This str should give absolute path and filename of granule for getting mask 
    filter_step: int 
        Horizontal step for sift data
    lonmin: float
        Minimum bounding box longitude
    lonmax: float
        Maximum bounding box longitude 
    latmin: float
        Minimum bounding box latitude
    latmax: float
        Maximum bounding box latitude 
    
    Returns
    -------
    land_mask : array
        array of land mask

    Examples
    ------
    >>> LAND_MASK = get_land_mask(latmin=60.0)
    ... plt.imshow(LAND_MASK)
    """
    data, error = get_data_from_ssmi_granule(filename, filter_step=4, init_land=True, 
                                                    lonmin=lonmin, lonmax=lonmax, 
                                                    latmin=latmin, latmax=latmax)
    return data['land']
```

## Базовые классы

**GranuleSSMIPhysical** - класс для описания гранул, считываемых с диска

```python
class GranuleSSMIPhysical(object):
    """Class describe SSMI snapshot files.
    attributes: satellite name, created time, time range type

    Examples:
    -------
    >>> g = GranuleSSMIPhysical('f10_19940709v7_d3d.gz', '/media/SOLabNFS/store/satellite/ssmi/f10/bmaps_v07/y1994/m07/')
    >>> g.print_grid_border()
    >>> g.print_polar_border()
    """
    boundary_points = []
    grid_step = .25
    
    def __init__(self, filename, path, root="/media/SOLabNFS/store/satellite/ssmi/", 
            is_weekly=False, initialize_data=False, create_border=False, step=4,
            latmin=60.0, latmax=90.0):
        
        self.filename = filename
        self.path = path
        self.root = root

        self.is_weekly = is_weekly
        self.range_type = "day"
        self.parse_name()
        
        self.verified = None
        
        self.latmin = latmin
        self.latmax = latmax
        
        self.y_latmax = 0
        self.y_latmin = (self.latmax - self.latmin) / self.grid_step
        self.step = step
        
        self.border_created = False
        self.data_initialized = False
        
        initialize_data and self.init_data()
        create_border and self.init_border()
    
    def __repr__(self):
        return "<SSMI granule: %s %s %s at %s>" % (self.range_type, self.filename, 
            'processed' if self.border_created else 'non-processed', id(self))
        
    def parse_name(self):
        """Initializing next attributes: satellite, range_type, created_at"""
        m = re.match(SSMI_FILENAME_RE, self.filename)
        #try:
        self.satellite_number = m.group('ss')

        year = int(m.group('year'))
        month = int(m.group('month'))
        day = 1 if m.group('day') is None else int(m.group('day'))

        self.created_at = datetime.date(year, month, day)
        self.year_and_week = (self.created_at.isocalendar()[0], self.created_at.isocalendar()[1])
        
        if self.is_weekly:
            self.range_type =  "week"
        elif m.group('day') is None and not m.group('date_type'):
            self.range_type = "month"
        elif not m.group('date_type'):
            self.range_type = "day" 
        elif m.group('date_type') == '_d3d':
            self.range_type = "3days"
    
    # data methods
    def init_data(self, lonmin=-0, lonmax=360, latmin=-90, latmax=90):
        self.data, self.error = get_data_from_ssmi_granule(os.path.join(self.path, self.filename), 
                lonmin, lonmax, latmin, latmax, filter_step=self.step)
        return self.data
     
    def clear_data(self):
        try: 
            delattr (self, "data")
        except AttributeError:
            pass
    
    # visualization methods
    def print_grid_border(self):
        self.border_created or self.init_border()
        fig = plt.figure(facecolor="w")
        plt.imshow(
            LAND_MASK * 10 + 
            self.data['ice']  
        )
        plt.axis([0, 1440 / self.step, 0, (self.latmax - self.latmin) / self.grid_step])
        plt.gca().invert_yaxis()
        plt.plot(self.gridded_boundary, color='#ffffff')
        self.clear_data()
        self.data_initialized = False
    
    # initialize border methods
    def init_border(self):
        self.init_data()
        self.create_border()
        self.clear_data()
    
    def print_polar_border(self):
        self.border_created or self.init_border()
        fig = plt.figure(figsize=(4,4),facecolor='#ffffff')
        ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)
        ax.plot(self.border_theta, self.boundary_points)
    
    # getter
    def __getattr__(self, key):
        if key == 'data':
            try: 
                self.__dict__[key]
            except KeyError:
                self.__dict__[key] = self.init_data(latmin=self.latmin)
                return self.__dict__[key]
            else: 
                return self.__dict__[key]
        if key in ['gridded_boundary', 'border_theta','polygon', 'area', 'square']:
            try: 
                self.__dict__[key]
            except KeyError:
                if not self.border_created:
                    self.init_border()
            else: 
                return self.__dict__[key]
        elif key == 'error': 
            try: 
                self.__dict__[key]
            except KeyError:
                self.init_data(latmin=self.latmin)
                return self.error
        else:
            return self.__dict__[key]
    
    def create_border(self):
        self.border_theta = self.data['lon'] * (np.pi / 180.0)  # polar degree
        self.xdim = len(self.border_theta)
        
        labeled_array, num_features = ndimage.label(self.data['ice'])  # get one big ice area
        data_ice = np.where(labeled_array == 1, 1, 0)
        
        
        #TODO: hardcode hack =)
        self.square = np.ma.array(SQUARES_MATRIX[0:95], mask=~self.data['ice'][0:95]+2).sum()
        
        # Create border of the ice with GRIDDED_MAX_BORDER filter
        self.gridded_boundary = np.zeros(self.xdim)
        for i in range(self.xdim):
            self.gridded_boundary[i] =  np.max(np.nonzero(data_ice[:,i] * self.data['lat']))
        
        self.gridded_boundary = np.amin(np.vstack([self.gridded_boundary, GRIDDED_MAX_BORDER]), axis=0)
        
        # translate grid into degrees
        self.boundary_points = [self.latmax - self.latmin - 
            (self.y_latmin - y) * self.grid_step 
            for y in self.gridded_boundary]
        
        #self.square = self.gridded_boundary.sum()
        
        return (self.border_theta, self.gridded_boundary, self.boundary_points)
```

Класс приложение **Arctic Ice** описывает основную бизнес логику и предоставляет следующие методы: инициализация данных, сканирование каталога, выборка данных, построение временных графиков, сохранение и загрузка данных

```python
class ArcticIceApplication(object):
    """ Class for ArcticIce application
    Examples
    -------
    >>> app = ArcticIceApplication()
    >>> app.registrationSSMI()
    """
    def __init__(self, ssmi_root = "/media/SOLabNFS/store/satellite/ssmi/", 
                       save_file="/home/nonamenix/notebooks/ssmiArctic.sv"):
        
        self.ssmi_root  = ssmi_root
        self.ssmi_granules = {
            'day': OOBTree.OOBTree(),
            '3days':OOBTree.OOBTree(),
            'week': OOBTree.OOBTree(),
            'month': OOBTree.OOBTree() }
        self.save_file = save_file
    
    def save_data(self):
        dump = shelve.open(self.save_file)
        for time_range in app.ssmi_granules:
            dump['ssmi_granules'][time_range] = self.ssmi_granules[time_range]
        dump.close()
    
    def load_data(self):
        dump = shelve.open(self.save_file)
        self.ssmi_granules = dump['ssmi_granules']
        dump.close()
    
    # Business logic 
    def registration_ssmi_granules(self):
        """Search granules in the root folder

        """
        os.chdir(self.ssmi_root)
        count = 0
        with Timer("Scanning ssmi folder"):
            for (path, dirs, files) in os.walk(self.ssmi_root):
                is_weekly = True if path.find('weeks') != -1  else False
                for filename in files:
                    if re.match(SSMI_FILENAME_RE, filename):
                        granule = GranuleSSMIPhysical(filename, path, self.ssmi_root, is_weekly)
                        self.ssmi_granules[granule.range_type][filename] = granule
                        count += 1
                        if count % 10000 == 0: 
                            print "- %d was counted" % count
        print "Total count of initialized snapshots: %d" % count
        print 'granules: ', { 
            'day':  len(self.ssmi_granules['day']),
            '3days': len(self.ssmi_granules['3days']),
            'week': len(self.ssmi_granules['week']),
            'month': len(self.ssmi_granules['month'])}

    # Error calculting
    def calculate_error(self, time_range='3days'):
        """ Function calculate error for granules and return array of them

        """
        i = 0 
        errors = []
        with Timer('Calculate error'):
            count = float(len(self.ssmi_granules[time_range]))
            for granule in self.ssmi_granules[time_range]: 
                errors.append(self.ssmi_granules[time_range][granule].error)
                self.ssmi_granules[time_range][granule].create_border()
                self.ssmi_granules[time_range][granule].clear_data()
                if i % 100 == 0:
                    print i, 
                i += 1 
        return errors
    
    # Block of calculating error distribution for satellites. Needed for verifing granules
    def calculate_errors_distibution_for_satellites(self):
        """Calculate statistic for granules grouped by satellites 

        """
        time_range = '3days'
        granules = [self.ssmi_granules[time_range][g] for g in self.ssmi_granules[time_range]]
        satellites = sorted(set([g.satellite_number for g in granules]))
        print 'satellites:', satellites
        description = dict()
        i = 0
        for ss in satellites:
            granules_grouped_by_satellite = sorted([g for g in granules if g.satellite_number == ss], key=lambda g: g.created_at)
            errors = [g.error for g in granules_grouped_by_satellite]
            d = sqrt(np.average([x**2 for x in errors]) - np.average(errors) ** 2)
            if ss == 'f15':  # hack :D
                d*=2
            description[ss] = {'dispersion':d, 'avg': np.average(errors)}
        
        print '\ndescription:'
        for key in description:
            print "%s: %s +/- %s" % (key, description[key]['avg'], description[key]['dispersion'])
        return time_range, granules, satellites, description
    
    def verify_granules(self, satellites, description, granules):
        for ss in satellites:
            avg = description[ss]['avg']
            desp = description[ss]['dispersion']
            granules_grouped_by_satellite = sorted([g for g in granules if g.satellite_number == ss], key=lambda g: g.created_at)
            for g in granules_grouped_by_satellite:
                g.verified = g.error < avg + desp and g.error > avg - desp
    
    def print_error_distibution_charts(self, granules, satellites, description):
        i = 0
        figures = dict()
        for ss in satellites:
            figures[i] = plt.figure(i)
            fig = figures[i]
            fig.set_facecolor('w')
            fig.set_size_inches(8, 4)
            fig.clear(); fig.clf()
            
            avg = description[ss]['avg']
            desp = description[ss]['dispersion']
            
            granules_grouped_by_satellite = sorted([g for g in granules if g.satellite_number == ss], key=lambda g: g.created_at)
            timeline = [g.created_at for g in granules_grouped_by_satellite]
            errors = [g.error for g in granules_grouped_by_satellite]
            plt.plot(timeline, errors, label=ss, color='#888888')
            
            filtered_granules_grouped_by_satellite = [g for g in granules_grouped_by_satellite if g.verified]
            timeline = [g.created_at for g in filtered_granules_grouped_by_satellite]
            errors = [g.error for g in filtered_granules_grouped_by_satellite]
            plt.plot(timeline, errors, 'b')
            plt.plot(timeline, [avg + desp for _ in range(len(timeline))], 'r--', alpha=.5)
            plt.plot(timeline, [avg - desp for _ in range(len(timeline))], 'r--', alpha=.5)
        
            filtered_granules_grouped_by_satellite = [g for g in granules_grouped_by_satellite if not g.verified]
            timeline = [g.created_at for g in filtered_granules_grouped_by_satellite]
            errors = [g.error for g in filtered_granules_grouped_by_satellite]
            plt.plot(timeline, errors, 'rx')
            print 'satellite: %s (%s/%s)' % (ss, len(filtered_granules_grouped_by_satellite), len(granules_grouped_by_satellite))
            i+=1
        
            plt.title('Error distribution for %s satellite' % ss)
            plt.legend()
            plt.tight_layout()
    
    @staticmethod
    def print_work_period_of_the_satellites(granules, satellites):
        mpl.rcParams['axes.color_cycle'] = COLORS
        i = 1
        fig = plt.figure(facecolor="w")
        fig.set_size_inches(8, 4)
        fig.clear()
        for ss in satellites:
            granules_grouped_by_satellite = sorted([g for g in granules if g.satellite_number == ss], key=lambda g: g.created_at)
            timeline = [g.created_at for g in granules_grouped_by_satellite]
            plt.plot(timeline, [ i for _ in range(len(timeline))],  label=ss, color=COLORS[i])
            plt.scatter([timeline[0], timeline[-1]], [i, i], s=50, zorder=2, marker='|', color=COLORS[i])
            i += 1
        plt.axvline(x=datetime.datetime.today(), ymin=0, ymax=1, c="g", linewidth=1, zorder=0, clip_on=False)
        plt.ylim([0, 9])
        plt.legend(loc=2)
        plt.tight_layout()
    
    def define_forecasting_granules(self):
        time_range = '3days'
        verified_granules = [self.ssmi_granules[time_range][g] for g in self.ssmi_granules[time_range] if self.ssmi_granules[time_range][g].verified]
        satellites = ['f08', 'f11', 'f15', 'f17']
        satellites_priority = {'f08':1, 'f11':2, 'f15':3, 'f17':4}
        granules = [g for g in verified_granules if g.satellite_number in satellites]
        forecasting_timeline = sorted(set([g.created_at for g in granules]))
        forecasting_granules = []
        with Timer('Forecasting granules'):
            counter = 0
            for date in forecasting_timeline:
                date_granules = [g for g in granules if g.created_at == date]
                if len(date_granules) == 1: 
                    forecasting_granules.append(date_granules[0])
                else:
                    forecasting_granules.append(max(date_granules, key=lambda g: satellites_priority[g.satellite_number]))
                counter += 1
                if counter % 1000 == 0: print counter, 
        return forecasting_granules
    
    @staticmethod
    def print_forecasting_granules_squares(forecasting_granules, satellites):
        fig = plt.figure(facecolor='#ffffff')
        fig.set_size_inches(8, 4)
        fig.clear()
        satellites = sorted(satellites)
        i = 0
        square_min = np.min([g.square for g in forecasting_granules])
        for ss in satellites:
            timeline = [g.created_at for g in forecasting_granules if g.satellite_number==ss]
            square = [g.square for g in forecasting_granules if g.satellite_number==ss]
            plt.plot(timeline, square, label=ss, color=COLORS[i])
            
            min_square = np.min(square)
            #line = square_min + 3000 - 1000 * (i % 4)
            plt.plot(timeline, [min_square for _ in range(len(timeline))], '--', color=COLORS[i])
            plt.scatter([timeline[0], timeline[-1]], [min_square, min_square], s=30, zorder=2, marker='|', color=COLORS[i])
            i += 1
        plt.legend(loc=3)
        plt.tight_layout()
        
    @staticmethod
    def init_ice_border_forecasting_granules(forecasting_granules):
        i = 0
        with Timer('Create border and calculate square for forecasting granules'):
            for granule in forecasting_granules:
                granule.init_border()
                if i % 100 == 0:
                    print i,
                i += 1
    
    @staticmethod
    def average_data_by_week(forecasting_granules):
        result_granules = []
        intervals = sorted(set(([g.year_and_week for g in forecasting_granules])))
        for i in intervals:
            granules = [g for g in forecasting_granules if g.year_and_week == i]
            fg = GranuleSSMIForecasting()
            fg.set_attributes_by_granules(granules)
            result_granules.append(fg)
        return result_granules
```
