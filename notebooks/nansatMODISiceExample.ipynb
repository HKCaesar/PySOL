```python
from nansat import Nansat
from domain import Domain

from osgeo import gdal
from numpy import array

import os
from sys import exit

import tempfile,logging

import urllib2, time
```

### Define functions for further Download

```python
def download(url, fout):
   """ Saves the url file to fout filename """
   filein  = urllib2.urlopen(url)
   fileout = open(fout, "wb")

   while True:
      bytes = filein.read(1024)
      fileout.write(bytes)

      if bytes == "": break

   filein.close()
   fileout.close()
```

### Initiate the Download part

```python
# ======== Change here =========
# Satellite name
# "aqua" or "terra"
MODIS_SATNAME = "terra"

# Subset name
# See the list on http://rapidfire.sci.gsfc.nasa.gov/subsets
SUBSET_NAME = "Arctic"
# The pixel size
# "2km", "1km", "500m" or "250m"
# NB! For Arctic miosaic: "4km", "2km", "1km",
RAPIDFIRE_RES = "1km"

# Year is in format "yyyy"
SUBSET_YEAR = str(time.localtime()[0])
# The day of the year
# SUBSET_CODE = "094"
SUBSET_CODE = "%03d" % (time.localtime()[7])
TUPLE_MODIS = (SUBSET_NAME, SUBSET_YEAR, SUBSET_CODE, MODIS_SATNAME, RAPIDFIRE_RES)
# The download urls
URL_RAPIDFIRE_SUBSET = "http://lance-modis.eosdis.nasa.gov/imagery/subsets/?mosaic=%s.%s%s.%s.%s.tif" % TUPLE_MODIS
IMAGE_FILE           = '%s.%s%s.%s.%s.tif' % TUPLE_MODIS
```

### Make tmp folder, specify Input/Output files

```python
# Make the output dir
oPath = '/media/SOLabNFS2/tmp/modis/arctic/'
figureNameTiff = oPath+IMAGE_FILE

if not os.path.exists(oPath):
    os.makedirs(oPath)

# Make a tmp dir
iPath = tempfile.mkdtemp()
fileName = os.path.join(iPath, IMAGE_FILE)

# Create Log File
# os.remove(oPath + 'procMODISmosaicArctic.log')
logging.basicConfig(filename=oPath + 'procMODISmosaicArctic.log', level=logging.INFO, format='%(asctime)s %(message)s', datefmt='%m/%d/%Y %I:%M:%S %p')

#~ Change the folder permissions
os.chmod(oPath, 0777)
```

```python
# Check if file already processed
if os.path.isfile(figureNameTiff):
    print "%s already processed" % (IMAGE_FILE)
    logging.info('%s already processed', IMAGE_FILE)
    #continue
else:
    print "%s" % (IMAGE_FILE)
```

### Main Download Part

```python
logging.info('%s Downloading true-color image from MODIS...', IMAGE_FILE)
download(URL_RAPIDFIRE_SUBSET, fileName)
```

```python
logging.info('%s Reading, reprojecting and exporting', IMAGE_FILE)
n = Nansat(fileName, mapperName='generic')
```

```python
m1 = n[1]
m2 = n[2]
m3 = n[3]
```

```python
mArray = ( (m1==255) == (m2==255) ) == (m3==255)
mArray2 = ( (m1==0) == (m2==0) ) == (m3==0)
```

```python
mArray = mArray + mArray2
mArray = mArray.astype('uint8')
del mArray2, m1, m2, m3
```

```python
n.add_band(bandID=4, array=mArray, \
   parameters={'name':'mask', \
   'wkv': 'mask', \
   'dataType': 2, 'minmax': '0 1'})
```

```python
lons, lats = n.get_geolocation_grids()
```

```python
plt.imshow(lats)
plt.colorbar()
```

```python
# Reproject the image
n.reproject()
pxlRes = array((1000, 1000))
srsString = 3413
d = Domain(srs=srsString, ds = n.vrt.dataset)
n.reproject(d)
```

```python
mArray = n[4]

# Write one band to GeoTIFF for geoserver using write_figure method
n.write_figure(figureNameTiff, bands=[1,2,3], clim='hist', mask_array=mArray, mask_lut={1: [0,0,0]})

# read the GeoTIFF image and save it with GDAL using LZW compression
dataset = gdal.Open(figureNameTiff, gdal.GA_ReadOnly)
band = dataset.GetRasterBand(1)
z = band.ReadAsArray()

# Remove tmp nonLZW GeoTIFF
os.remove(figureNameTiff)

# Write out the LZW GeoTIFF 
out = gdal.GetDriverByName('Gtiff').Create(figureNameTiff, \
                                                          band.XSize, \
                                                          band.YSize, 1, \
                                                          gdal.GDT_Byte, \
                                                          ['COMPRESS=LZW'])
out.GetRasterBand(1).WriteArray(z)
out = None
# Add proj param to GeoTIFF
n.vrt.copyproj(figureNameTiff)
```
