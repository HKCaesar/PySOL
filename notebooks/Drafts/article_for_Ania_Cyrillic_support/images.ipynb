```python
import matplotlib.pyplot as plt
import scipy.io
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
```

```python
%matplotlib inline
path = '/home/monzik/article/fonts/arial.ttf'
prop = font_manager.FontProperties(fname=path)
prop.set_size(24)
mpl.rcParams['font.family'] = prop.get_name()
mpl.rcParams['xtick.labelsize'] = 20
mpl.rcParams['ytick.labelsize'] = 20
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['grid.linewidth'] = 1.5
```

```python
mat = scipy.io.loadmat('py02Jan2012_0833_001.mat')
```

```python
plt.close('all')
plt.figure(figsize=(10,10), facecolor='w')
plt.plot(mat['x1'][0], mat['y1'][0],linewidth = 2, color='black', linestyle='-')
plt.plot(mat['x2'][0], mat['y2'][0],linewidth = 2, color='black', linestyle='--', dashes=[10,7])
plt.plot(mat['x3'][0], mat['y3'][0],linewidth = 2, color='black', linestyle='--', dashes=[10,6,2,6])
plt.xlabel(u'Расстояние от берега, км', fontproperties=prop)
plt.ylabel(ur'Ветровое поверхностное напряжение, $\mathrm{Н/м^2}$', fontproperties=prop)
plt.grid(True)

l = plt.legend([u'РСА',ur'модельные, изменение''\n'ur'$\mathrm{z_0}$ и $\mathrm{\theta}$ поверхности, U=%s м/с'%mat['U'][0][0], 
           ur'модельные, изменение''\n'ur'$\mathrm{z_0}$ поверхности, U=%s м/с'% mat['U2'][0][0]],  loc=4, prop=prop,framealpha=0.99)
plt.xlim(0,mat['xl'])
plt.ylim(0,1)
plt.savefig('02Jan2012_0833_001.tif', format='tif', dpi=600)
```
