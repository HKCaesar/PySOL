```python
# Helper functions and classes
class Timer(object):
    """
    Timer — counting the elapsed spent.
    >>> with Timer("create ice boundary"):
    >>>     sime_func()...
    """
    def __init__(self, name=None):
        self.name = name

    def __enter__(self):
        self.tstart = time.time()

    def __exit__(self, type, value, traceback):
        if self.name:
            print '[%s]' % self.name,
        print 'Elapsed: %s seconds' % (time.time() - self.tstart)
        
def simplify_granule(granule):
    simple = dict.fromkeys(['created_at', 'border_theta', 'boundary_points', 'gridded_boundary', 'square', 'satellite_number'])
    for attr in simple.keys():
        simple[attr] = granule.__dict__[attr]
    return simple

def get_season(date):
  # convert date to month and day as integer (md), e.g. 4/21 = 421, 11/17 = 1117, etc.
  m = date.month * 100
  d = date.day
  md = m + d
  if ((md > 320) and (md < 621)):
    s = 'spring' #spring
  elif ((md > 620) and (md < 923)):
    s = 'summer' #summer
  elif ((md > 922) and (md < 1223)):
    s = 'fall' #fall
  else:
    s = 'winter' #winter
  return s
```

```python
class Granule(object):
    def __init__(self, simple_granule=dict(), granules=False):
        if len(simple_granule.keys()) > 0:
            for attr in simple_granule.keys():
                self.__dict__[attr] = simple_granule[attr]
        
        if granules:
            self.average(granules)
        
        self.enrich_meta_information()
        
    def __repr__(self):
        return "<SSMI granule %s - %s at %d>" % (self.created_at.strftime('%Y.%m.%d'), self.ended_at.strftime('%Y.%m.%d'), id(self))
    
    def average(self, granules):
        granules = sorted(granules, key=lambda g: g.created_at)
        self.created_at = (min(granules, key=lambda g: g.created_at)).created_at
        self.ended_at = max(granules, key=lambda g: g.created_at).created_at
        
        self.border_theta = granules[0].border_theta
        self.satellite_number = granules[0].satellite_number
        
        # Calculate average ice border
        borders = [g.boundary_points for g in granules]
        self.boundary_points = np.mean(np.vstack(borders), axis=0)
        self.gridded_boundary = [int(round(p * 4)) for p in self.boundary_points]
        self.square = np.mean([g.square for g in granules])
    
    def enrich_meta_information(self):
        self.week = self.created_at.isocalendar()[1]
        self.year = self.created_at.year
        self.month = self.created_at.month
        self.season = get_season(self.created_at)
        
        self.year_week = "%s-%s" % (self.year, self.week)
        self.year_month = "%s.%s" % (self.year, self.month)
        self.year_season = "%s %s" % (self.year, self.season)
    
    def print_polar_chart(self):
        fig = plt.figure(figsize=(4,4),facecolor='#ffffff')
        ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)
        ax.plot(self.border_theta, self.boundary_points, label="%s" % self.created_at)
        ax.fill_between(self.border_theta, 0, self.boundary_points, alpha=.1)
        ax.legend()
```

```python
class GranulesCollection(list):
    def __init__(self, granules):
        self.collection = sorted(granules, key=lambda g: g.created_at)
        self.calculate_statistic()
    
    def __len__(self):
        return self.collection.__len__()
    
    def __getitem__(self, key):
        return self.collection.__getitem__(key)
    
    def __setitem__(self, key, value):
        self.collection.__setitem__(key, value)
    
    def __repr__(self):
        return "<SSMI list with %s granules at %s>" % (len(self), id(self))
    
    def __iter__(self):
        return self.collection.__iter__()
    
    def __getslice__(self, i, j):
        return GranulesCollection(self.collection.__getslice__(i, j))
    
    def calculate_statistic(self):
        self.timeline = [g.created_at for g in self.collection]
    
    def get_time_series_by_meridians(self, meridians):
        meridians = dict.fromkeys(meridians)
        for m in meridians:
            meridians[m] = [g.boundary_points[m] for g in self.collection]
        return meridians
    
    def print_time_series_by_meridians(self, meridians, step=4):
        granules = self.collection[::step]
        count_all = len(meridians)
        fig = plt.figure(facecolor='w')
        fig.set_size_inches(8, count_all * 2)
        fig.clear()
        axes = dict.fromkeys(meridians)
        timeline = [g.created_at for g in granules]
        meridians_data = dict.fromkeys(meridians)
        i = 1
        for m in meridians:
            subplot = count_all * 100 + 10 + i
            axes[m] = fig.add_subplot(subplot)
            data = [g.boundary_points[m] for g in granules]
            
            min_max = [np.min(data) - 2, np.max(data) + 2]
            axes[m].plot(timeline, data, label=u"%s° meridian" % (m), color=COLORS[i])
            axes[m].set_ylim(min_max)
            axes[m].fill_between(timeline, min_max[0], data, alpha=.1, color=COLORS[i])
            axes[m].legend()
            axes[m].grid()
            i += 1
        plt.tight_layout()
    
    def print_polar_chart(self, fill=0, show_max=False):
        mpl.rcParams['axes.color_cycle'] = COLORS
        fig = plt.figure(figsize=(6,6),facecolor='#ffffff')
        ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True)
        i = 0
        for g in self.collection:
            ax.plot(g.border_theta, g.boundary_points, label="%s-%s (%s) " % (g.created_at.strftime('%Y%m%d'), g.week, int(g.square)), color = COLORS[i], zorder=i)
            ax.fill_between(g.border_theta, 0, g.boundary_points, alpha=fill, color = COLORS[i], zorder=i)
            i+=1
        
        if show_max:
            ax.plot(g.border_theta, MAX_BORDER, "r-", label="MAX_BORDER", linewidth=.5, zorder=0)
            ax.fill_between(g.border_theta, 0, MAX_BORDER, alpha=.05, color="r", zorder=0)
            
        ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),  ncol=3, fancybox=True, fontsize=8)
        #ax.tight_layout()
        
    def query(self, weeks=False, months=False, years=False, seasons=False):
        def time_range(t):
            return [t] if type(t) == int else t

        weeks = time_range(weeks)
        months = time_range(months)
        years = time_range(years)
        seasons = time_range(seasons)

        return GranulesCollection([g for g in self.collection if 
            (not weeks or g.week in weeks) and 
            (not years or g.created_at.year in years) and 
            (not months or g.created_at.month in months) and
            (not seasons or g.season in seasons) ])
    
    def average_granules(self, by='year_week'):
        # by: week, year_week, month, year, season
        result_granules = []
        intervals = sorted(set(([getattr(g, by) for g in self.collection])))
        for i in intervals:
            granules = [g for g in self.collection if getattr(g, by) == i]
            granules = sorted(granules, key=lambda g: g.created_at)
            result_granules.append(Granule(granules=granules))
        return GranulesCollection(result_granules)
```

```python
def print_ice_square_chart_light_by_years(self, light=True, show_min=True, show_max=False, show_avg=False, show_years=True, colors_count = 4):
    fig = plt.figure(facecolor='#ffffff')
    fig.set_size_inches(12, 4)
    fig.clear()
    years = sorted(set([g.created_at.year for g in self.collection]))
    i = 0
    min_square_year_granules = []
    max_square_year_granules = []
    avg_timeline = []
    avg_squares = []
    min_square_granule = min(self.collection, key=lambda g: g.square)
    
    squares = [g.square for g in self.collection]
    plt.plot(self.timeline, squares)
    
    for year in years:
        year_granules = [g for g in self.collection if g.created_at.year==year]
        timeline = [g.created_at for g in year_granules]
        squares = [g.square for g in year_granules]
        plt.plot(timeline, squares, color=COLORS[i % colors_count])
       
        if show_min:
            min_square_year_granules.append(min(year_granules, key=lambda g: g.square))
        
        if show_max:
            max_square_year_granules.append(max(year_granules, key=lambda g: g.square))
        
        if show_avg: 
            year_len = len(timeline)
            avg_timeline.append(timeline[year_len/2])
            avg_squares.append(np.mean(squares))
        
        if show_years: 
            plt.fill_between(timeline, min_square_granule.square, squares, alpha=.1, color=COLORS[i % colors_count])
        
        if light:
            i += 1
            
    if show_min:
        timeline = [g.created_at for g in min_square_year_granules]
        squares = [g.square  for g in min_square_year_granules]
        plt.plot(timeline, squares, label='min')
    
    if show_max:
        timeline = [g.created_at for g in max_square_year_granules]
        squares = [g.square  for g in max_square_year_granules]
        plt.plot(timeline, squares, label='max', color=COLORS[4])
    
    if show_avg: 
        plt.plot(avg_timeline, avg_squares, label="average", linewidth=2.0, alpha=.5)
    
    plt.grid(True)
    plt.legend(loc=3)
    plt.tight_layout()

GranulesCollection.print_ice_square_chart_light_by_years = print_ice_square_chart_light_by_years
```
