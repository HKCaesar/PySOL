```python
from nansat import Nansat, Domain
import numpy as np
import matplotlib.pyplot as plt
```

```python
xdim = 1440
ydim = 720

data = {}.fromkeys(['lon', 'lat', 'lon_grid', 'lat_grid', 'image'])
data['lon']=np.zeros((xdim))
data['lat']=np.zeros((ydim))

for i in xrange(xdim):
    data['lon'][i] = 0.25 * (i + 1) - 0.125 
for y in xrange(ydim):
    data['lat'][y]=0.25*(y+1) - 90.125

data['lon_grid'], data['lat_grid'] = \
    np.meshgrid(data['lon'],data['lat'],)

# flip the lat_grid/land/ice so that North would be +90 at top
data['lat_grid'] = data['lat_grid'][::-1]
```

```python
# Generate Some data
data['image'] =  sqrt(data['lat_grid']**2 + data['lon_grid']**2) + sin(data['lat_grid']**2 + data['lon_grid']**2)
plt.close('all')
plt.figure(figsize=(17,2))
plt.subplot(131)
plt.imshow(data['image'])
plt.colorbar()
plt.subplot(132)
plt.imshow(data['lat_grid'])
plt.colorbar()
plt.subplot(133)
plt.imshow(data['lon_grid'])
plt.colorbar()
```

```python
gridDomain = Domain(lon=data['lon_grid'], lat=data['lat_grid'])
grid = Nansat(domain=gridDomain, array=data['image'])
grid.add_band(array=data['lat_grid'])
grid.add_band(array=data['lon_grid'])

plt.close('all')
plt.figure(figsize=(5,2))
plt.imshow(grid[1])
plt.colorbar()
```

```python
grid.vrt.add_geolocationArray()
lons, lats = grid.get_geolocation_grids()

print (lats - data['lat_grid']).all() == 0
print '\n'
print (lons - data['lon_grid']).all() == 0
```

```python
pt1 = (80.0,60.0) #(lon, lat)
pt2 = (80.0,0.0)
pt3 = (80.0,0.0)
pt4 = (80.0,-60.0)

grid_trn, llVec  = grid.get_transect((pt1, pt2, pt3, pt4), bandList=[1], latlon=True) # transect, [lonVector, latVector]

plt.close('all')
plt.figure(figsize=(17,3))
plt.subplot(131)
plt.plot(llVec[0])

plt.subplot(132)
plt.plot(llVec[1])

plt.subplot(133)
plt.plot(grid_trn[0])
```
