```python
from nansat import Nansat
from domain import Domain

from osgeo import gdal
from numpy import array

import os, errno, time, urllib2
from sys import exit

import tempfile,logging
```

```python
def silentrm(filename):
    try:
        os.remove(filename)
    except OSError as e: # this would be "except OSError as e:" in python 3.x
        if e.errno != errno.ENOENT: # errno.ENOENT = no such file or directory
            raise # re-raise exception if a different error occured
```

```python
def download(url, fout):
   """ Saves the url file to fout filename """
   filein  = urllib2.urlopen(url)
   fileout = open(fout, "wb")

   while True:
      bytes = filein.read(1*1024*1024) # 1Mbytes
      fileout.write(bytes)

      if bytes == "": break

   filein.close()
   fileout.close()
```

```python
def jpgRead(fn, vrt):
    """ Create Nansat object from jpg file and specified vrt """
    # Open correct tif and export vrt
    # n = Nansat('/media/SOLabNFS2/tmp/modis/' + SUBSET_NAME + '.2013009.terra.' + RAPIDFIRE_RES + '.tif', mapperName='generic')
    # n.vrt.export('/media/SOLabNFS2/tmp/modis/mosaic' + SUBSET_NAME + 'VRT.xml')
    
    # Create empty Nansat object with Arctic.tif vrt
    n = Nansat(vrt)
    # Create domain
    d = Domain(ds=n.vrt.dataset)
    # Read as array Arctic.jpg when can't read tif
    from PIL import Image
    im=Image.open(fn)
    im = array(im.convert('RGB'))
    # Create Nansat object with Arctic.jpg and domain "d"
    n = Nansat(array=im[:,:,0], domain=d)
    n.add_band(bandID=2, array=im[:,:,1])
    n.add_band(bandID=3, array=im[:,:,2])
    return n
```
