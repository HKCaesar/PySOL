```python
#  ----------------------------------
# Parallel version takes longer
# 1 loops, best of 1: 53.5 s per loop
# %%timeit -n 1 -r 1

# INTERPOLATE DATA
# Interpolate Geolocation grid points and calibration LUTs onto grid of raw_counts.shape

from multiprocessing import Process, Queue
from scipy.interpolate import RectBivariateSpline

def RectBivariateSpline_que(line, pixel, data, line_2, pixel_2, num2split, que=[]):
    data_2 = RectBivariateSpline(line, pixel, data, kx=2, ky=2)(line_2, pixel_2)
    
    # split large array
    data_2 = array_split(data_2, num2split)
    ## Write to the queue
    for ii in xrange(0, num2split):
        que.put(data_2[ii])


spline = {} # empty dict for interpolated values
q = {}      # empty dict for queue
num2split = 20

for qq in xrange(0,5):
    q[str(qq)] = Queue()
    spline[str(qq)] = []

# Interpolate lat/lon onto a new grid
p0 = Process(target=RectBivariateSpline_que, \
                args=(GEOgrid['line'][:,0], \
                      GEOgrid['pixel'][0,:], \
                      GEOgrid['lats'], \
                      line_2, pixel_2, num2split, q['0']))
p1 = Process(target=RectBivariateSpline_que, \
                args=(GEOgrid['line'][:,0], \
                      GEOgrid['pixel'][0,:], \
                      GEOgrid['lons'], \
                      line_2, pixel_2, num2split, q['1']))
# interpolate incidenceAngle
p2 = Process(target=RectBivariateSpline_que, \
                args=(GEOgrid['line'][:,0], \
                      GEOgrid['pixel'][0,:], \
                      GEOgrid['incidenceAngle'], \
                      line_2, pixel_2, num2split, q['2']))
# interpolate sigmaNought
p3 = Process(target=RectBivariateSpline_que, \
                args=(cLUTs['line'][:,0], \
                      cLUTs['pixel'][0,:], \
                      cLUTs['sigmaNought'], \
                      line_2, pixel_2, num2split, q['3']))
# interpolate noiseLut
p4 = Process(target=RectBivariateSpline_que, \
                args=(nLUTs['line'][:,0], \
                      nLUTs['pixel'][0,:], \
                      nLUTs['noiseLut'], \
                      line_2, pixel_2, num2split, q['4']))

for qq in xrange(0,5):
    eval('p' + str(qq) + '.start()')

for qq in xrange(0,5):
    for ii in xrange(0, num2split):
        spline[str(qq)].append(q[str(qq)].get())

for qq in xrange(0,5):
    eval('p' + str(qq) + '.join()')

lats_2 = concatenate(spline['0'])#.astype(GEOgrid['lats'].dtype)
lons_2 = concatenate(spline['1'])
incidenceAngle_2 = concatenate(spline['2'])
sigmaNought_2 = concatenate(spline['3'])
noiseLut_2 = concatenate(spline['4'])
```
